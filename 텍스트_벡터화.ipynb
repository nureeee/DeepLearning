```python
# n번 문서에 단어가 등장한 횟수
def term_frequency(term, document):
     tf = document.count(term) # term이 들어간 횟수
     
     return tf

def document_frequency(term, documents):
    df = 0
    for i in range(len(documents)):
        if term in documents[i]:
            df = df + 1

    # 방법2
    # df += term in documents 하면 값이 True와 False로 나와 True면 1이 된다.
    #방법3
    # for document in documents:
    #     if term in document:
    #         df += 1

    return df

def inverse_document_frequency(term, documents):
    from math import log
    # len(set(' '.join(documents).split())) n을 구하긴 한건데 형태소 분리가 되지 않은 것이라 무시
    idf = log(len(documents) / (1 + document_frequency(term, documents)))

    return idf

def tf_idf(term, documents, idx):
    result = term_frequency(term, documents[idx]) * inverse_document_frequency(term, documents)
    
    return result
```
