# 전체 페이지 크롤링
페이지 넘기는 것이 get방식이었다면 주소가 바뀌는데 바뀌지 않는다는 것은 get이 아닌 다른방식  
network에서 확인해보자 (머리글(head) 확인)
> post방식으로 넘긴는 것을 알 수 있다,.


```python
last_page = 102
```

```python
url = 'https://www.gangwon.to/gwtour/gangwondo_trip/tourist_spot'
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
datas = []

for page in range(1, last_page+1):
    #1. 서버에 전송할 데이터 준비
    data = {'page index': page}
    # 이렇게만 넘기는 방식을 form-data
    # json으로 넘겨야 하면.. content-type: json이면
    # jsom.dumps(data) 감싸서 전송
    
    response = requests.post(url, data).content
    soup = BeautifulSoup(response, 'html.parser')

    total_ul = soup.select_one("#contents > div.contsArea.skinContainer > ul.nlist01.clearfix")
    all_li   = total_ul.select("li")

    for li in all_li:
        text_div = li.select_one("div.text")
        category = text_div.select_one('span').text
        title = text_div.select_one('strong').text
        desc = text_div.select_one('p').text

        scrap_data = {
        '카테고리' : category,
        '관광지면' : title,
        '관광지 소개' : desc
        }

        datas.append(scrap_data)
```

```python
tour_df = pd.DataFrame(datas)
tour_df.head()
```
